```python
import os
from PIL import Image

Br = Image.open("BracketBase.png")
Av = Image.open("AvalancheCircle.png")
RX = Image.open("RedX.png")

Av.paste(RX,(0,0),RX)
display(Av)

bw,bh = Br.size
xw,xh = RX.size

Br.paste(Av,(int(bw/2-xw/2),0))
Br.paste(RX,(int(bw/2-xw/2),0),RX)

if not os.path.exists("NewFolder"): os.mkdir("NewFolder")
Br.save("NewFolder/ShowingTyler.png", "PNG")
display(Br)
```
